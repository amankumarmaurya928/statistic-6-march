```python
# Q5
'''
1. Nominal: the data can only be categorized.
           1. Quantitative/ categorical variable.
           2. order does't matter.
           ex: color, gender , labedls

2. Ordinal: the data can be categorized and ranked.
            1. ranking and order matter
            2. difference can't be measured.
            ex: ranking(like on basis of makes in exam)

3. Interval: the data can be categorized and ranked, and evenly spaced.
            1. ranking and order matter.
            2. difference can be measured but not ratio.
            3. does't have zero as starting value.
            ex: temperature(it may be negative also)

4. Ratio: the data can be categorized, ranked, evenly spaced and has a natural zero.
             1. rank and order matters.
             2. difference and ratio can be measured.
             3. zero is starting point.
             ex: grades(or scores)
'''
```
